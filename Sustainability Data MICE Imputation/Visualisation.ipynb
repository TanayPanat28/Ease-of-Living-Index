```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_sus = pd.read_csv("Sustainability_index.csv")

selected_countries = ["India", "United States", "China", "Australia", "Malaysia", "Canada", "Germany", "United Kingdom", "Japan"]

filtered_df = df_sus[df_sus['Country'].isin(selected_countries)]

avg_renewable_excluding_hydro_selected = filtered_df.groupby('Country')['Electricity production_renewable'].mean()
avg_renewable_energy_consumption_selected = filtered_df.groupby('Country')['Greenhouse_emission'].mean()


plt.figure(figsize=(10, 6))
avg_renewable_energy_consumption_selected.sort_values().plot(kind='bar', color='green')
plt.title('Green House Gas Emission')
plt.ylabel('Emission')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
selected_countries = ["India", "United States", "China", "Australia", "Malaysia", "Canada", "Germany", "United Kingdom", "Japan"]

filtered_df = df_sus[df_sus['Country'].isin(selected_countries)]

# Calculate average values for renewable energy consumption and greenhouse emission
avg_renewable_excluding_hydro_selected = filtered_df.groupby('Country')['Electricity production_renewable'].mean()
avg_renewable_energy_consumption_selected = filtered_df.groupby('Country')['Greenhouse_emission'].mean()

# Sorting the data for better readability
avg_renewable_energy_consumption_selected = avg_renewable_energy_consumption_selected.sort_values()

# Plotting as a horizontal bar chart using plt.barh()
plt.figure(figsize=(10, 6))
plt.barh(avg_renewable_energy_consumption_selected.index, avg_renewable_energy_consumption_selected.values, color='lightgreen', edgecolor='black', height=0.7)  # Increased bar height

# Title and axis labels
plt.suptitle('Greenhouse Gas Emission', fontsize=14, weight='bold', ha='left', x=0.13)
plt.text(0.0, 1.02, "The emission is in metric tonnes.", fontsize=12, style='italic', ha='left', transform=plt.gca().transAxes)

plt.xlabel('Emission')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Adding text boxes inside each bar
for index, value in enumerate(avg_renewable_energy_consumption_selected):
    plt.text(value - 1.5, index, f'{value:.2f}', va='center',weight='bold', ha='right', fontsize=10, color='black', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Filter the data to include only years from 2000 to 2020
filtered_df = filtered_df[(filtered_df['Year'] >= 2000) & (filtered_df['Year'] <= 2020)]

# Define colors for highlighted countries
highlighted_countries = {
    "Germany": "black", 
    "United Kingdom": "blue", 
    "India": "orange", 
    "China": "red", 
    "Malaysia": "brown"
}
bright_colors = ["purple", "cyan", "magenta", "teal", "lime", "pink", "navy", "coral"]  # Colors for other countries
color_index = 0

# Plot each country's data
for country in selected_countries:
    df_country = filtered_df[filtered_df['Country'] == country]
    if country in highlighted_countries:
        plt.plot(df_country['Year'], df_country['Sustainability_index_FA'], label=country, linewidth=2.5, color=highlighted_countries[country])
        # Adding text for the latest year's index value
        latest_year = df_country['Year'].iloc[-1]
        latest_value = df_country['Sustainability_index_FA'].iloc[-1]
        plt.text(latest_year + 0.1, latest_value, f'{latest_value:.2f}', fontsize=10, color=highlighted_countries[country], va='center')
    else:
        # Use dashed lines and cycle through bright colors
        plt.plot(df_country['Year'], df_country['Sustainability_index_FA'], label=country, linewidth=1.5, linestyle='--', color=bright_colors[color_index])
        color_index = (color_index + 1) % len(bright_colors)  # Cycle through bright colors

plt.xlabel('Year', fontsize=12)
plt.ylabel('Sustainability Index', fontsize=12)
plt.title('Sustainability Index', fontsize=14, weight='bold', ha='left', x=0)  # Align the title to the left
# Set x-ticks at 5-year intervals
plt.xticks(range(2000, 2021, 5), fontsize=10)
plt.legend(title="Country", loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Filter the data to include only years from 2000 to 2020
filtered_df = filtered_df[(filtered_df['Year'] >= 2000) & (filtered_df['Year'] <= 2020)]

# Define colors for highlighted countries
highlighted_countries = {"Germany": "black", "United Kingdom": "blue", "India": "orange", "China": "red", "Malaysia": "brown"}
default_color = "grey"

# Plot each country's data
for country in selected_countries:
    df_country = filtered_df[filtered_df['Country'] == country]
    if country in highlighted_countries:
        plt.plot(df_country['Year'], df_country['Sustainability_index_FA'], label=country, linewidth=2.5, color=highlighted_countries[country])
        # Adding text for the latest year's index value
        latest_year = df_country['Year'].iloc[-1]
        latest_value = df_country['Sustainability_index_FA'].iloc[-1]
        plt.text(latest_year + 0.1, latest_value, f'{latest_value:.2f}', fontsize=10, color=highlighted_countries[country], va='center')
    else:
        plt.plot(df_country['Year'], df_country['Sustainability_index_FA'], label=country, linewidth=1.5, color=default_color, alpha=0.6)

plt.xlabel('Year', fontsize=12)
plt.ylabel('Sustainability Index', fontsize=12)
plt.title('Sustainability Index', fontsize=14, weight='bold', ha='left', x=0)  # Align the title to the left
# Set x-ticks at 5-year intervals
plt.xticks(range(2000, 2021, 5), fontsize=10)  # Ensure x-ticks are set to integers at 5-year intervals
plt.legend(title="Country", loc='upper left')
plt.tight_layout()
plt.show()
```
